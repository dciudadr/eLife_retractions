```python
import os
import sys
from ftplib import FTP
from fnmatch import fnmatch
from urllib.parse import urlparse
from tqdm import tqdm

def download_ftp_file(ftp_client, filename, output_filename):
    # use temp files so we know the file is not complete
    temp_output_filename = output_filename + '.part'
    if os.path.isfile(output_filename):
        os.remove(output_filename)
    with open(temp_output_filename, 'wb') as f:
        # need to change the type to binary for the size command to succeed
        ftp_client.voidcmd('TYPE I')
        total = ftp_client.size(filename)

        with tqdm(total=total, unit='B', unit_scale=True, leave=False, file=sys.stdout) as pbar:
            pbar.set_description(filename)

            def cb(data):
                pbar.update(len(data))
                f.write(data)

            ftp_client.retrbinary('RETR {}'.format(filename), cb)
    os.rename(temp_output_filename, output_filename)

def download_files_from_ftp(ftp_url, file_pattern, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    parsed_ftp_url = urlparse(ftp_url)
    with FTP(parsed_ftp_url.hostname) as ftp_client:
        ftp_client.login()
        ftp_client.cwd(parsed_ftp_url.path)
        filenames = ftp_client.nlst()
        matching_filenames = [fn for fn in filenames if fnmatch(fn, file_pattern)]
        for filename in matching_filenames:
            output_filename = os.path.join(output_dir, filename)
            if os.path.isfile(output_filename):
                print('already downloaded:', output_filename)
            else:
                download_ftp_file(ftp_client, filename, output_filename)
                print('downloaded:', output_filename)

def download_file_from_ftp(ftp_url, output_dir):
    download_files_from_ftp(
        os.path.dirname(ftp_url),
        os.path.basename(ftp_url),
        output_dir
    )

    
data_path = 'data'

# articles
articles_output_dir = os.path.join(data_path, 'FTP_PUBMED_papers')
articles_ftp_url = 'ftp://ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/'
articles_file_pattern = '*.xml.tar.gz'
# Or you could start by just downloading one file
#file_pattern = 'non_comm_use.A-B.xml.tar.gz'

# pmc ids
pmc_ids_output_dir = os.path.join(data_path, 'PMC_ids')
pmc_ids_ftp_url = 'ftp://ftp.ncbi.nlm.nih.gov/pub/pmc/PMC-ids.csv.gz'
data_path = 'data'
pmc_ids_output_dir = os.path.join(data_path, 'PMC_ids')
pmc_ids_ftp_url = 'ftp://ftp.ncbi.nlm.nih.gov/pub/pmc/PMC-ids.csv.gz'
download_file_from_ftp(pmc_ids_ftp_url, pmc_ids_output_dir)


# oa_file_list.csv 
data_path = 'data'
oa_file_list_output_dir = os.path.join(data_path, 'oa_file_list')
oa_file_list_ftp_url = 'ftp://ftp.ncbi.nlm.nih.gov/pub/pmc/oa_file_list.csv'
download_file_from_ftp(oa_file_list_ftp_url, oa_file_list_output_dir)
```

# Let's find the tarfiles to check and get the text

```python
import string

def list_tarfiles_to_check (initial_word):
    
    initial_word = initial_word.lower()
    initial_letter = initial_word[0]
    letter_index = string.ascii_lowercase.index(initial_letter)

    # A-B
    if letter_index <= 1:
        files_to_check = ["comm_use.A-B.xml.tar.gz", "non_comm_use.A-B.xml.tar.gz"]
    # C-H
    if (2 <= letter_index) & (letter_index <= 7):
        files_to_check = ["comm_use.C-H.xml.tar.gz", "non_comm_use.C-H.xml.tar.gz"]

    # I-N
    if (8 <= letter_index) & (letter_index <= 13):
        files_to_check = ["comm_use.I-N.xml.tar.gz", "non_comm_use.I-N.xml.tar.gz"]

    # O-Z
    if (14 <= letter_index) & (letter_index <= 25):
        files_to_check = ["comm_use.O-Z.xml.tar.gz", "non_comm_use.O-Z.xml.tar.gz"]

    return files_to_check

                               
assert list_tarfiles_to_check("A Journal") == ["comm_use.A-B.xml.tar.gz", "non_comm_use.A-B.xml.tar.gz"]
assert list_tarfiles_to_check("C Journal") == ["comm_use.C-H.xml.tar.gz", "non_comm_use.C-H.xml.tar.gz"]
assert list_tarfiles_to_check("J Journal") == ["comm_use.I-N.xml.tar.gz", "non_comm_use.I-N.xml.tar.gz"]
assert list_tarfiles_to_check("S Journal") == ["comm_use.O-Z.xml.tar.gz", "non_comm_use.O-Z.xml.tar.gz"]
```

```python
import tarfile
import re
from datetime import datetime

def get_text_from_tar_file(PMID, list_tar_files):

    startTime = datetime.now()
    print("start = ", startTime)
          
    for file in list_tar_files:
        file_path = '/project/data/' + file
        print(file_path)

        tar = tarfile.open(file_path, "r:gz")
          
        for filename in tar:
            if re.search(PMID, filename.name) != None:
                print(filename.name)
                paper = tar.extractfile(filename)
                #print(paper.read())
                #tar.close()
                return paper.read()
        tar.close()
        
        print("end = ",  datetime.now())
        print("total = ",  datetime.now() - startTime )
         
    return "Article not found"
```
