Anlyse infomrmation about papers and retractions directly taken from this webpage. The search details were:

The results displayed were saved as retractions600.txt

```python
import pandas as pd

df = pd.read_csv('retractions600.txt', sep = '\t')

df = df.reset_index()
df
```

```python
df2 = df[['level_0','Date/PubMedID/DOI', 'Date/PubMedID/DOI.1']]
df2 = df2.dropna(subset = ['Date/PubMedID/DOI'])
df2 = df2.dropna().reset_index(drop = 'True')
df2
```

```python
df2.shape[0] % 3
```

```python
from datetime import datetime
from dateutil.parser import parse
import pandas as pd

DOI_article = []
PubMed_ID_article = []
DOI_retraction = []
PubMed_ID_retraction = []
#Journal = []
date_article = []
date_retraction = []

for index, row in df2.iterrows():
    if len(row['Date/PubMedID/DOI']) > 10: 
        try:
            date_article.append(datetime.strptime(df2.iloc[index -2, 1], '%m/%d/%Y'))
            date_retraction.append(datetime.strptime(df2.iloc[index -2, 2], '%m/%d/%Y'))

            DOI_article.append(row['Date/PubMedID/DOI'])
            DOI_retraction.append(row['Date/PubMedID/DOI.1'])
        
            PubMed_ID_article.append(df2.iloc[index -1, 1])
            PubMed_ID_retraction.append(df2.iloc[index -1, 2])
            
        except ValueError:
            pass
                      
        
        #print(row['Date/PubMedID/DOI'],row['Date/PubMedID/DOI.1'])

# df2 = pd.DataFrame({'Journal': Journal,'DOI':DOI_a
df_retracted_retraction = pd.DataFrame({'doi':DOI_article, 
                    'pmid_article':PubMed_ID_article, 
                    'doi_retraction':DOI_retraction, 
                    'pmid_retraction':PubMed_ID_retraction,
                    'date_article': date_article,
                    'date_retraction' : date_retraction}).drop_duplicates()

#mask = df_retracted_retraction.date_retraction != 'unavailable'
#mask4 = df_retracted_retraction.date_retraction != "0"
#df_retracted_retraction = df_retracted_retraction[mask3 & mask4]
df_retracted_retraction

```
