```python
import tarfile
import re
from datetime import datetime

def get_text_from_tar_file(PMID, list_tar_files):

    startTime = datetime.now()
    print("start = ", startTime)
          
    for file in list_tar_files:
        file_path = '/project/data/' + file
        print(file_path)

        tar = tarfile.open(file_path, "r:gz")
          
        for filename in tar:
            if re.search(PMID, filename.name) != None:
                print(filename.name)
                paper = tar.extractfile(filename)
                print(paper.read())
                tar.close()
                return
        tar.close()
        
        print("end = ",  datetime.now())
        print("total = ",  datetime.now() - startTime )
         
          
```

```python
import string

def list_tarfiles_to_check (initial_word):
    
    initial_word = initial_word.lower()
    initial_letter = initial_word[0]
    letter_index = string.ascii_lowercase.index(initial_letter)

    # A-B
    if letter_index <= 1:
        files_to_check = ["comm_use.A-B.xml.tar.gz", "non_comm_use.A-B.xml.tar.gz"]
    # C-H
    if (2 <= letter_index) & (letter_index <= 7):
        files_to_check = ["comm_use.C-H.xml.tar.gz", "non_comm_use.C-H.xml.tar.gz"]

    # I-N
    if (8 <= letter_index) & (letter_index <= 13):
        files_to_check = ["comm_use.I-N.xml.tar.gz", "non_comm_use.I-N.xml.tar.gz"]

    # O-Z
    if (14 <= letter_index) & (letter_index <= 25):
        files_to_check = ["comm_use.O-Z.xml.tar.gz", "non_comm_use.O-Z.xml.tar.gz"]

    return files_to_check

                               
assert list_tarfiles_to_check("A Journal") == ["comm_use.A-B.xml.tar.gz", "non_comm_use.A-B.xml.tar.gz"]
assert list_tarfiles_to_check("C Journal") == ["comm_use.C-H.xml.tar.gz", "non_comm_use.C-H.xml.tar.gz"]
assert list_tarfiles_to_check("J Journal") == ["comm_use.I-N.xml.tar.gz", "non_comm_use.I-N.xml.tar.gz"]
assert list_tarfiles_to_check("S Journal") == ["comm_use.O-Z.xml.tar.gz", "non_comm_use.O-Z.xml.tar.gz"]
```
